## Part 3: Model Fitting (8 points)

### Task A: Define Your Physics Model

Based on **your assigned experiment**, define the appropriate model function:

---

#### **Pressure (Elevator) → Barometric Formula**

Pressure vs altitude:
$$
P(h) = P_0 e^{-h/H}
$$

where $P_0$ is sea-level pressure, $H \approx 8500$ m is scale height.

Convert time to height using $h = v \cdot t$ (elevator speed $v \approx 1$ m/s).

**Fit:** `P(t) = P0 * exp(-v*t / H)`

**Report:** $P_0$ and $v$ (elevator speed)

---

#### **Gyroscope (Chair) → Exponential Damping**

Angular velocity decays due to friction:
$$
\omega(t) = \omega_0 e^{-\beta t}
$$

**Fit:** `omega(t) = omega0 * exp(-beta * t)`

**Report:** $\omega_0$ (initial angular velocity) and $\beta$ (damping constant)

---

#### **Magnetometer (Magnet) → Dipole Field**

Magnetic field from a dipole:
$$
B(r) = \frac{\mu_0 m}{4\pi r^3}
$$

Simplify: $B(r) = A / r^3$

You need to convert time to distance (measure your hand speed separately, or fit $r(t) = r_0 + v t$).

**Fit:** `B(t) = A / (r0 + v*t)**3`

**Report:** $A$ (dipole strength) and $v$ (approach speed)

---

#### **Microphone (Sound Decay) → Exponential Damping**

Sound amplitude decays:
$$
A(t) = A_0 e^{-\gamma t}
$$

**Fit:** `A(t) = A0 * exp(-gamma * t)`

**Report:** $A_0$ (initial amplitude) and $\gamma$ (decay rate)

```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

def dipole_model(t, A, v):
    return A / (0.5 + v*t)**3

time = df_trim.iloc[:,0].values
B_field = df_trim.iloc[:,1].values


popt, pcov = curve_fit(dipole_model, time, B_field,)
A_fit, v_fit = popt
uncertainties = np.sqrt(np.diag(pcov))
A_err, v_err = uncertainties


plt.figure(figsize=(8, 5))
plt.scatter(time, B_field, label='Data', alpha=0.6)
plt.plot(time, dipole_model(time, *popt), 'r-', label='Fit', linewidth=2)
plt.xlabel('Time (s)')
plt.ylabel('Magnetic Field (B)')
plt.legend()
plt.title('Magnetometer: Dipole Field Fit')
plt.show()

# Print fitted parameters with uncertainties
print(f"A = {A_fit:.3f} ± {A_err:.3f}")
print(f"v = {v_fit:.3f} ± {v_err:.3f} m/s")
```

### Deliverable

```python
B_fit = dipole_model(time, *popt)
residuals = B_field - B_fit

fig, axs = plt.subplots(2, 2, figsize=(12, 8))

axs[0,0].plot(time, B_field, 'b.', alpha=0.6)
axs[0,0].set_title("Raw Data")
axs[0,0].set_xlabel("Time (s)")
axs[0,0].set_ylabel("Magnetic Field (B)")
axs[0,0].grid(True)

axs[0,1].plot(time, B_field, 'b.', alpha=0.6, label="Data")
axs[0,1].plot(time, B_fit, 'r-', linewidth=2, label="Fit")
axs[0,1].set_title("Fitted Dipole Model")
axs[0,1].set_xlabel("Time (s)")
axs[0,1].set_ylabel("Magnetic Field (B)")
axs[0,1].legend()
axs[0,1].grid(True)


axs[1,0].plot(time, residuals, 'k.', alpha=0.6)
axs[1,0].axhline(0, color='r', linestyle='--', linewidth=1)
axs[1,0].set_title("Residuals (Data - Fit)")
axs[1,0].set_xlabel("Time (s)")
axs[1,0].set_ylabel("Residual (B)")
axs[1,0].grid(True)

axs[1,1].hist(residuals, bins=20, color='gray', edgecolor='black', alpha=0.7)
axs[1,1].set_title("Histogram of Residuals")
axs[1,1].set_xlabel("Residual (B)")
axs[1,1].set_ylabel("Frequency")
axs[1,1].grid(True)


plt.tight_layout()
plt.show()
```
