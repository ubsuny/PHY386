```python
def string_operations(Galaxies):
  print("Lowercase:", Galaxies.lower())
  print("Uppercase:", Galaxies.upper())
  print("Capitalize:", Galaxies.capitalize)
  print("Reverse:",Galaxies[::-1])
  print("Length:", len(Galaxies))
```

```python
string_operations("Galaxies")
```
