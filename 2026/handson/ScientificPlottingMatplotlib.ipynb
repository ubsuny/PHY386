## Introduction and Imports

A plot is not decoration — it is an argument. In experimental physics, a figure communicates your measurement, its uncertainty, and your interpretation all at once. Matplotlib is the standard plotting library in Python and gives you precise control over every element of a figure.

The two standard imports you will use in almost every notebook:

````python
import numpy as np
import matplotlib.pyplot as plt

# Set 11 pt fonts globally — matches typical publication text size
plt.rcParams.update({
    'font.size': 11,
    'axes.labelsize': 11,
    'axes.titlesize': 11,
    'xtick.labelsize': 11,
    'ytick.labelsize': 11,
    'legend.fontsize': 11,
})
````

## Basic Line and Scatter Plots

We start with projectile motion. A ball is launched at angle $\theta = 45°$ with initial speed $v_0 = 20$ m/s. The trajectory follows:

$$y(x) = x \tan\theta - \frac{g x^2}{2 v_0^2 \cos^2\theta}$$

`plt.plot()` connects points with a line; `plt.scatter()` draws individual markers. Use `plt.show()` at the end of each cell to display the figure.

````python
g = 9.81          # m/s^2
v0 = 20.0         # m/s
theta = np.radians(45)

x = np.linspace(0, v0**2 * np.sin(2 * theta) / g, 300)
y = x * np.tan(theta) - (g * x**2) / (2 * v0**2 * np.cos(theta)**2)

plt.plot(x, y)
plt.xlabel("Horizontal distance (m)")
plt.ylabel("Height (m)")
plt.title("Projectile Trajectory")
plt.show()
````

`plt.scatter()` is preferred when each data point is an independent measurement (e.g., from an experiment) rather than a continuous function. Here we simulate discrete time-sampled positions of the same projectile:

````python
t_samples = np.linspace(0, 2 * v0 * np.sin(theta) / g, 15)
x_samples = v0 * np.cos(theta) * t_samples
y_samples = v0 * np.sin(theta) * t_samples - 0.5 * g * t_samples**2

plt.scatter(x_samples, y_samples, color="steelblue")
plt.xlabel("Horizontal distance (m)")
plt.ylabel("Height (m)")
plt.title("Sampled Projectile Positions")
plt.show()
````

## Labels, Titles, and Units

**Always label your axes with the physical quantity and its unit.** A plot without units is incomplete — a reader cannot extract any physical meaning from it. Use `plt.xlabel()`, `plt.ylabel()`, and `plt.title()`. Units go in parentheses at the end of the axis label: `"Force (N)"`, `"Time (s)"`.

The following example shows Hooke's Law, $F = kx$, with $k = 5$ N/m:

````python
k = 5.0  # spring constant, N/m
displacement = np.linspace(0, 0.2, 50)  # meters
force = k * displacement                 # Newtons

plt.plot(displacement, force, color="crimson")
plt.xlabel("Displacement (m)")
plt.ylabel("Force (N)")
plt.title(r"Hooke's Law: $F = kx$,  $k = 5$ N/m")
plt.show()
````

## LaTeX in Matplotlib

Matplotlib renders LaTeX math inside any string that contains `$...$`. To avoid Python interpreting backslashes, always use **raw strings**: `r'$\alpha$'` instead of `'$\\alpha$'`.

Common physics symbols and their raw strings:

| Symbol | Raw string |
|--------|------------|
| $\alpha, \beta, \gamma$ | `r'$\alpha$'`, `r'$\beta$'`, `r'$\gamma$'` |
| $\omega, \lambda, \Delta$ | `r'$\omega$'`, `r'$\lambda$'`, `r'$\Delta$'` |
| $v$ (m/s) | `r'$v$ (m/s)'` |
| $E = mc^2$ | `r'$E = mc^2$'` |
| $\hbar\omega$ | `r'$\hbar\omega$'` |

````python
omega = 2 * np.pi * 1.0   # angular frequency, rad/s
t = np.linspace(0, 2, 500)
amplitude = 0.05           # meters
displacement_sho = amplitude * np.cos(omega * t)

plt.plot(t, displacement_sho)
plt.xlabel(r"Time $t$ (s)")
plt.ylabel(r"Displacement $x(t)$ (m)")
plt.title(r"Simple Harmonic Oscillator: $x(t) = A\cos(\omega t)$")
plt.show()
````

You can also add LaTeX annotations directly onto the plot with `plt.text()`. This is useful for labeling a specific feature, such as the amplitude or frequency:

````python
plt.plot(t, displacement_sho)
plt.xlabel(r"Time $t$ (s)")
plt.ylabel(r"Displacement $x(t)$ (m)")
plt.title("Simple Harmonic Oscillator")
plt.text(0.05, 0.042, r"$A = 5$ cm", fontsize=11, color="steelblue")
plt.text(0.05, 0.035, r"$\omega = 2\pi$ rad/s", fontsize=11, color="steelblue")
plt.show()
````

## Figure Size and DPI

`plt.figure(figsize=(width, height), dpi=N)` controls the canvas size (in inches) and resolution (dots per inch). These two settings have very different purposes:

- `figsize` determines the **aspect ratio and physical size** of the rendered figure
- `dpi` determines **pixel density** — higher values mean sharper output

**For publications, use `figsize=(6, 4)`.** A US letter page is 8.5" wide. With 1" margins on each side, your text column is 6.5" wide. A 6" figure fits comfortably within that column. The 3:2 aspect ratio (6x4) is a good default for most physics plots.

**The font size in your figure should match the font size in your document.** If your paper uses 11 pt text, your axis labels and tick labels should also be ~11 pt. Otherwise the figure will look out of place when embedded. Use `dpi=300` when saving for a publication. Screen display typically uses 100 dpi.

````python
# Wide figure suits time-series data — e.g., temperature vs. time
time_series = np.linspace(0, 10, 200)   # seconds
temperature = 300 + 15 * np.sin(0.8 * time_series) + np.random.normal(0, 1, 200)

plt.figure(figsize=(6, 3), dpi=100)
plt.plot(time_series, temperature, linewidth=0.8)
plt.xlabel("Time (s)")
plt.ylabel("Temperature (K)")
plt.title("Thermal Oscillation — wide figure for time series")
plt.tight_layout()
plt.show()
````

## Customizing Line Styles and Markers

Matplotlib gives you fine-grained control over line appearance. The most useful keyword arguments are:

- `linestyle`: `'-'`, `'--'`, `':'`, `'-.'`
- `linewidth`: float, in points
- `marker`: `'o'`, `'s'`, `'^'`, `'x'`, `'+'`, `'D'`
- `markersize`: float
- `color`: named color `'crimson'`, hex `'#1f77b4'`, or RGB tuple

The example below compares three models for the same damped oscillation data so each curve needs a distinct, distinguishable style.

````python
t = np.linspace(0, 5, 300)
gamma_values = [0.3, 0.8, 1.5]
styles = [
    {"linestyle": "-",  "linewidth": 2.0, "color": "steelblue",  "label": r"$\gamma = 0.3$ s$^{-1}$"},
    {"linestyle": "--", "linewidth": 2.0, "color": "crimson",    "label": r"$\gamma = 0.8$ s$^{-1}$"},
    {"linestyle": ":",  "linewidth": 2.5, "color": "darkorange", "label": r"$\gamma = 1.5$ s$^{-1}$"},
]

plt.figure(figsize=(6, 4))
for gamma, style in zip(gamma_values, styles):
    plt.plot(t, np.exp(-gamma * t) * np.cos(2 * np.pi * t), **style)

plt.xlabel(r"Time $t$ (s)")
plt.ylabel(r"Amplitude $A(t)$")
plt.title("Damped Oscillations for Different Damping Coefficients")
plt.legend()
plt.show()
````

When plotting **discrete measurements** (not a continuous model), always use markers without a connecting line, or use `linestyle='none'` combined with a marker. Connecting experimental data points with lines implies continuity that may not exist.

````python
wavelengths = np.array([400, 450, 500, 550, 600, 650, 700])  # nm
intensities = np.array([0.12, 0.45, 0.78, 0.95, 0.82, 0.53, 0.21])  # normalized

plt.figure(figsize=(6, 4))
plt.plot(wavelengths, intensities, marker='D', linestyle='none',
         markersize=7, color='darkorchid')
plt.xlabel(r"Wavelength $\lambda$ (nm)")
plt.ylabel("Normalized Intensity")
plt.title("Emission Spectrum — Discrete Measurements")
plt.show()
````

## Color Choices for Science

**Avoid the `jet` colormap.** It has poor perceptual uniformity — equal steps in data produce unequal steps in perceived color — and it is completely unreadable when printed in greyscale. Use `viridis` (default since Matplotlib 2.0), `plasma`, `cividis` (colorblind-safe), or `RdBu` for diverging data.

For multi-line plots with a discrete number of curves, the tab10 palette (`C0`, `C1`, `C2`, ...) is a good starting point. It was designed to be distinguishable and is the Matplotlib default color cycle.

````python
# 2D interference pattern from a double slit
x = np.linspace(-5, 5, 300)
y = np.linspace(-5, 5, 300)
X, Y = np.meshgrid(x, y)
d = 1.5   # slit separation in arbitrary units
intensity_2d = np.cos(np.pi * d * X / 3)**2 * np.exp(-0.1 * (X**2 + Y**2))

fig, axes = plt.subplots(1, 2, figsize=(10, 4))
im0 = axes[0].imshow(intensity_2d, cmap='jet',    origin='lower', extent=[-5, 5, -5, 5])
im1 = axes[1].imshow(intensity_2d, cmap='viridis', origin='lower', extent=[-5, 5, -5, 5])
axes[0].set_title("jet — avoid this")
axes[1].set_title("viridis — use this")
for ax in axes:
    ax.set_xlabel(r"$x$ (a.u.)")
    ax.set_ylabel(r"$y$ (a.u.)")
fig.colorbar(im1, ax=axes[1], label="Intensity (a.u.)")
plt.tight_layout()
plt.show()
````

For data with a meaningful zero — like electric or magnetic fields that can be positive or negative — use a **diverging colormap** such as `RdBu` (red-white-blue). The white center corresponds to zero, making it immediately obvious where the field changes sign. This is standard in E&M visualizations.

````python
# Electric potential from two point charges (dipole)
x = np.linspace(-3, 3, 300)
y = np.linspace(-3, 3, 300)
X, Y = np.meshgrid(x, y)

# Positive charge at (-1, 0), negative charge at (+1, 0)
r_pos = np.sqrt((X + 1)**2 + Y**2)
r_neg = np.sqrt((X - 1)**2 + Y**2)
V = np.where((r_pos > 0.15) & (r_neg > 0.15),
             1.0 / r_pos - 1.0 / r_neg, np.nan)

fig, ax = plt.subplots(figsize=(6, 5), layout='constrained')
im = ax.imshow(V, cmap='RdBu', origin='lower', extent=[-3, 3, -3, 3],
               vmin=-3, vmax=3)
fig.colorbar(im, ax=ax, label=r"Potential $V$ (a.u.)")
ax.set_xlabel(r"$x$ (a.u.)")
ax.set_ylabel(r"$y$ (a.u.)")
ax.set_title(r"Electric Dipole Potential — RdBu diverging colormap")
ax.plot(-1, 0, 'ko', markersize=8)  # positive charge
ax.plot(1, 0, 'ko', markersize=8)   # negative charge
ax.text(-1, -0.4, "+", ha='center', fontsize=14, fontweight='bold')
ax.text(1, -0.4, "−", ha='center', fontsize=14, fontweight='bold')
plt.show()
````

## Error Bars

Error bars are not optional in experimental physics — they are how you communicate the uncertainty of each measurement. Use `plt.errorbar()`. The `yerr` argument accepts a scalar (same error on every point) or an array (individual errors).

The example below shows a measurement of gravitational acceleration $g$ at different drop heights. The uncertainty decreases with height because the relative timing uncertainty shrinks.

````python
height = np.array([0.20, 0.40, 0.60, 0.80, 1.00, 1.20])        # meters
g_measured = np.array([9.74, 9.83, 9.79, 9.81, 9.76, 9.82])     # m/s^2
g_uncertainty = np.array([0.15, 0.10, 0.08, 0.07, 0.06, 0.05])  # m/s^2

plt.figure(figsize=(6, 4))
plt.errorbar(height, g_measured, yerr=g_uncertainty,
             fmt='o', color='steelblue', ecolor='steelblue',
             capsize=5, capthick=1.5, linewidth=1.5,
             label="Measured $g$")
plt.axhline(9.81, color='crimson', linestyle='--', label=r"$g = 9.81$ m/s$^2$")
plt.xlabel("Drop Height (m)")
plt.ylabel(r"Measured $g$ (m/s$^2$)")
plt.title("Free-Fall Measurement of Gravitational Acceleration")
plt.legend()
plt.show()
````

`capsize` adds the horizontal end-caps on the error bars (in points). `fmt='o'` sets the data point marker. `ecolor` controls the error bar color independently of the data point color — useful when you want subtle grey error bars on a colored scatter.

**The `fmt` string uses the same syntax as `plt.plot()`:** `'s--'` gives square markers with a dashed line, `'none'` gives error bars only with no marker.

## Log Scales

Many physical relationships are power laws or exponentials: radioactive decay, signal attenuation, stellar luminosity functions. Log scales let you see these structures clearly.

- `plt.yscale('log')` — logarithmic y-axis
- `plt.xscale('log')` — logarithmic x-axis
- Both together give a **log-log plot** where a power law $y = ax^n$ appears as a straight line with slope $n$.

$$N(t) = N_0 \, e^{-\lambda t}$$

````python
decay_constant = 0.35   # s^-1 (hypothetical isotope)
N0 = 1000               # initial number of nuclei
t = np.linspace(0, 15, 200)
N_t = N0 * np.exp(-decay_constant * t)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

ax1.plot(t, N_t, color='steelblue')
ax1.set_xlabel(r"Time $t$ (s)")
ax1.set_ylabel(r"$N(t)$ (nuclei)")
ax1.set_title("Linear Scale")

ax2.plot(t, N_t, color='steelblue')
ax2.set_yscale('log')
ax2.set_xlabel(r"Time $t$ (s)")
ax2.set_ylabel(r"$N(t)$ (nuclei)")
ax2.set_title("Logarithmic y-axis — decay appears linear")

plt.tight_layout()
plt.show()
````

````python
# Luminosity vs. surface temperature (simplified Stefan-Boltzmann scaling)
# L ~ T^4 for blackbody stars
temperature_star = np.logspace(3.5, 5, 200)  # Kelvin
luminosity = (temperature_star / 5778)**4     # normalized to solar luminosity

plt.figure(figsize=(6, 4))
plt.loglog(temperature_star, luminosity, color='darkorange', linewidth=2)
plt.xlabel(r"Surface Temperature $T$ (K)")
plt.ylabel(r"Luminosity $L/L_\odot$")
plt.title(r"Stefan-Boltzmann Scaling: $L \propto T^4$")
plt.gca().invert_xaxis()  # HR diagram convention: hot stars on left
plt.show()
````

`plt.loglog()` is a shorthand for setting both axes to log scale at once. Notice the `invert_xaxis()` trick — the Hertzsprung-Russell diagram convention places hot (blue) stars on the left and cool (red) stars on the right.

## Multiple Plots: Subplots

`plt.subplots(nrows, ncols)` returns a figure and an array of `Axes` objects. Always use the **object-oriented interface** (`ax.set_xlabel(...)` rather than `plt.xlabel(...)`) when working with subplots — it makes it unambiguous which panel you are targeting.

The code below shows position, velocity, and acceleration for a mass on a spring, which are related by differentiation and naturally belong side by side.

````python
omega0 = 2.0   # rad/s
t = np.linspace(0, 4, 500)
A = 0.1        # amplitude in meters

x_pos = A * np.cos(omega0 * t)
x_vel = -A * omega0 * np.sin(omega0 * t)
x_acc = -A * omega0**2 * np.cos(omega0 * t)

fig, axes = plt.subplots(3, 1, figsize=(6, 7), sharex=True)

axes[0].plot(t, x_pos, color='steelblue')
axes[0].set_ylabel(r"$x$ (m)")
axes[0].set_title("Mass on a Spring")

axes[1].plot(t, x_vel, color='crimson')
axes[1].set_ylabel(r"$\dot{x}$ (m/s)")

axes[2].plot(t, x_acc, color='darkorange')
axes[2].set_ylabel(r"$\ddot{x}$ (m/s$^2$)")
axes[2].set_xlabel(r"Time $t$ (s)")

plt.tight_layout()
plt.show()
````

`sharex=True` makes all panels share the same x-axis so tick labels only appear on the bottom panel — a standard pattern for stacked physics plots.

## Legends and Annotations

`plt.legend()` reads the `label=` keyword from each plotted object. Place the legend with `loc=`: `'upper right'` (default), `'lower left'`, `'best'` (Matplotlib chooses), or outside the axes with `bbox_to_anchor`.

`plt.annotate()` draws an arrow pointing to a specific data feature — ideal for marking a resonance peak, a phase transition, or a maximum.

````python
omega = np.linspace(0.1, 5, 500)
omega0, gamma_r = 2.0, 0.3

def amplitude_resonance(w, w0, g):
    """Driven harmonic oscillator amplitude response."""
    return 1.0 / np.sqrt((w0**2 - w**2)**2 + (g * w)**2)

A_underdamped = amplitude_resonance(omega, omega0, gamma_r)
A_overdamped  = amplitude_resonance(omega, omega0, 1.5)

plt.figure(figsize=(6, 4))
plt.plot(omega, A_underdamped, label=r"$\gamma = 0.3$ (underdamped)", color='steelblue')
plt.plot(omega, A_overdamped,  label=r"$\gamma = 1.5$ (overdamped)",  color='crimson', linestyle='--')

# Annotate the resonance peak
peak_idx = np.argmax(A_underdamped)
plt.annotate(r"Resonance $\omega \approx \omega_0$",
             xy=(omega[peak_idx], A_underdamped[peak_idx]),
             xytext=(3.0, 2.5),
             arrowprops=dict(arrowstyle='->', color='black'),
             fontsize=11)

plt.xlabel(r"Driving frequency $\omega$ (rad/s)")
plt.ylabel(r"Amplitude $A(\omega)$ (a.u.)")
plt.title("Driven Harmonic Oscillator — Resonance")
plt.legend(loc='upper right')
plt.show()
````

## Grid, Ticks, and Spines

Grids and tick marks help a reader extract numerical values from a plot. The key methods are:

- `ax.grid(True, which='both', linestyle=':', alpha=0.5)` — enable grid
- `ax.tick_params(which='both', direction='in', top=True, right=True)` — inward ticks on all sides (publication style)
- `ax.set_xticks([...])` / `ax.set_yticks([...])` — explicit tick positions
- `ax.minorticks_on()` — enable minor tick marks

**Physics journals almost universally require inward ticks on all four sides.**

````python
wavelength_nm = np.linspace(380, 780, 500)  # visible spectrum range
# Simulated spectral intensity — three emission peaks
spectral_intensity = (
    0.6 * np.exp(-((wavelength_nm - 436)**2) / (2 * 8**2)) +   # blue mercury line
    0.9 * np.exp(-((wavelength_nm - 546)**2) / (2 * 8**2)) +   # green mercury line
    0.75 * np.exp(-((wavelength_nm - 615)**2) / (2 * 10**2))   # red neon line
)

fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(wavelength_nm, spectral_intensity, color='darkviolet', linewidth=1.5)
ax.set_xlabel(r"Wavelength $\lambda$ (nm)")
ax.set_ylabel("Intensity (a.u.)")
ax.set_title("Emission Spectrum — Tick and Grid Customization")

ax.set_xticks(np.arange(380, 800, 40))
ax.minorticks_on()
ax.tick_params(which='both', direction='in', top=True, right=True)
ax.grid(True, which='major', linestyle=':', alpha=0.5)
plt.tight_layout()
plt.show()
````

## Saving Figures for Publications

Use `fig.savefig()` immediately before `plt.show()`. The file format is inferred from the extension.

- **PDF or SVG** — vector formats, infinitely scalable, ideal for papers
- **PNG at 300 dpi** — acceptable for theses and conference slides
- **Never submit a JPEG** — lossy compression creates visible artifacts around text and lines

**Always include `bbox_inches='tight'`** — without it, axis labels outside the default bounding box are silently cropped.

````python
fig, ax = plt.subplots(figsize=(6, 4))
t = np.linspace(0, 2 * np.pi, 300)
ax.plot(t, np.sin(t), label=r"$\sin(\theta)$", color='steelblue')
ax.plot(t, np.cos(t), label=r"$\cos(\theta)$", color='crimson', linestyle='--')
ax.set_xlabel(r"Phase $\theta$ (rad)")
ax.set_ylabel("Amplitude")
ax.set_title("Trigonometric Functions")
ax.legend()

# Save as vector PDF for a paper
fig.savefig("trig_functions.pdf", bbox_inches='tight')

# Save as high-res PNG for a thesis
fig.savefig("trig_functions.png", dpi=300, bbox_inches='tight')

plt.show()
print("Figures saved: trig_functions.pdf and trig_functions.png")
````

**Tip:** save the figure object as `fig` from `plt.subplots()` and call `fig.savefig()` rather than `plt.savefig()`. This guarantees you save exactly the figure you built, even in notebooks that display intermediate plots in earlier cells.

## Advanced Tips and Tricks

This section collects practical patterns that come up repeatedly in physics research plotting. Each tip is self-contained.

### Twin Axes: Two y-Scales on One Plot

`ax.twinx()` creates a second y-axis sharing the same x-axis. Use this when two physically related quantities have different units or very different magnitudes — for example, plotting both voltage and current over the same time series in an RC circuit.

````python
t = np.linspace(0, 5e-3, 500)   # 5 ms
RC = 1e-3                        # time constant, seconds
V_in = 5.0                       # volts
voltage = V_in * (1 - np.exp(-t / RC))
current = (V_in / 100) * np.exp(-t / RC)  # 100 Ohm resistor

fig, ax1 = plt.subplots(figsize=(6, 4))
ax1.plot(t * 1e3, voltage, color='steelblue', label="Voltage")
ax1.set_xlabel("Time (ms)")
ax1.set_ylabel(r"Voltage $V$ (V)", color='steelblue')
ax1.tick_params(axis='y', labelcolor='steelblue')

ax2 = ax1.twinx()
ax2.plot(t * 1e3, current * 1e3, color='crimson', linestyle='--', label="Current")
ax2.set_ylabel(r"Current $I$ (mA)", color='crimson')
ax2.tick_params(axis='y', labelcolor='crimson')

plt.title("RC Circuit Charging — Twin y-Axes")
plt.tight_layout()
plt.show()
````

### Inset Plots

An inset plot lets you zoom into a region of interest without creating a separate panel. Use `ax.inset_axes([x0, y0, width, height])` where coordinates are in axes-fraction units (0 to 1). This is common in papers to show a detail of a resonance peak, a phase transition, or a fit residual.

````python
omega = np.linspace(0.5, 4.0, 1000)
A = 1.0 / np.sqrt((2.0**2 - omega**2)**2 + (0.2 * omega)**2)

fig, ax = plt.subplots(figsize=(6, 4))
ax.plot(omega, A, color='steelblue', linewidth=1.5)
ax.set_xlabel(r"$\omega$ (rad/s)")
ax.set_ylabel(r"Amplitude $A(\omega)$")
ax.set_title("Resonance with Inset Zoom")

# Inset: zoom into the peak region
ax_inset = ax.inset_axes([0.55, 0.45, 0.40, 0.45])
ax_inset.plot(omega, A, color='steelblue', linewidth=1.5)
ax_inset.set_xlim(1.7, 2.3)
ax_inset.set_ylim(0, A.max() * 1.1)
ax_inset.set_xlabel(r"$\omega$ (rad/s)", fontsize=8)
ax_inset.tick_params(labelsize=7)
ax.indicate_inset_zoom(ax_inset, edgecolor='grey')
plt.show()
````

### `rcParams` — Global Default Settings

`matplotlib.rcParams` is a dictionary of all default settings. You can override any of them at the top of a notebook to apply consistently across all figures. This is far preferable to repeating `fontsize=12` in every single call.

````python
plt.rcParams.update({
    'font.size': 12,
    'axes.labelsize': 13,
    'axes.titlesize': 13,
    'xtick.labelsize': 11,
    'ytick.labelsize': 11,
    'lines.linewidth': 1.8,
    'figure.dpi': 100,
    'savefig.dpi': 300,
    'savefig.bbox': 'tight',
})

# All subsequent figures now inherit these defaults
t = np.linspace(0, 2 * np.pi, 300)
plt.plot(t, np.sin(t))
plt.xlabel(r"Phase $\theta$ (rad)")
plt.ylabel(r"$\sin\theta$")
plt.title("After rcParams Update — note larger fonts")
plt.show()
````

**Tip:** reset to defaults at any time with `plt.rcParams.update(plt.rcParamsDefault)`. Place your `rcParams.update()` call in the same import cell at the top of the notebook so it applies globally from the start.

### Reusable Plot Functions

If you make the same type of plot many times (e.g., plotting a spectrum, a trajectory, or a time series), write a function for it. This avoids copy-paste errors and makes your analysis code much more readable.

````python
def plot_spectrum(wavelength, intensity, title="Spectrum", color='steelblue'):
    """
    Plot a single optical spectrum.

    Parameters
    ----------
    wavelength : array-like
        Wavelength array in nanometers.
    intensity : array-like
        Spectral intensity (arbitrary units).
    title : str, optional
        Figure title.
    color : str, optional
        Line color.
    """
    fig, ax = plt.subplots(figsize=(6, 4))
    ax.plot(wavelength, intensity, color=color, linewidth=1.5)
    ax.set_xlabel(r"Wavelength $\lambda$ (nm)")
    ax.set_ylabel("Intensity (a.u.)")
    ax.set_title(title)
    ax.minorticks_on()
    ax.tick_params(which='both', direction='in', top=True, right=True)
    plt.tight_layout()
    return fig, ax

# Use it:
wl = np.linspace(400, 700, 400)
spec = np.exp(-((wl - 532)**2) / (2 * 15**2))  # 532 nm laser line
fig, ax = plot_spectrum(wl, spec, title="Nd:YAG Laser — 532 nm Line", color='limegreen')
plt.show()
````

### `constrained_layout` vs `tight_layout`

Both prevent axis labels from being clipped, but they work differently:

- `plt.tight_layout()` — a post-hoc adjustment called after all plotting is done. Simple, always works.
- `fig, ax = plt.subplots(layout='constrained')` — a newer engine that adjusts dynamically as you add elements. Handles colorbars and `suptitle` better.

**Use `layout='constrained'` when you have colorbars, figure-level titles (`fig.suptitle`), or complex subplot grids.**

````python
# Magnetic field from a current loop — 2D cross section
x = np.linspace(-2, 2, 200)
y = np.linspace(-2, 2, 200)
X, Y = np.meshgrid(x, y)
r = np.sqrt(X**2 + Y**2)
B_field = np.where(r > 0.1, 1.0 / r**2, np.nan)  # avoid singularity at origin

fig, ax = plt.subplots(figsize=(6, 5), layout='constrained')
im = ax.imshow(B_field, cmap='plasma', origin='lower',
               extent=[-2, 2, -2, 2], vmax=5)
fig.colorbar(im, ax=ax, label=r"$|\mathbf{B}|$ (a.u.)")
ax.set_xlabel(r"$x$ (a.u.)")
ax.set_ylabel(r"$y$ (a.u.)")
ax.set_title("Magnetic Field Cross-Section (layout='constrained')")
plt.show()
````

## Summary and Best Practices

A checklist for every figure you produce in this course and beyond:

**Data integrity**
- Always show error bars on experimental measurements
- Never connect discrete data points with a line unless the underlying function is continuous
- Do not use `jet` — use `viridis`, `plasma`, or `cividis`; use `RdBu` for diverging data (e.g., E&M fields)

**Labels and units**
- Every axis must have a label that includes the physical quantity and its unit: `"Force (N)"`
- Use raw strings `r'$...$'` for LaTeX in any label, title, or annotation
- Add a title that describes the physics, not just the variables

**Layout and appearance**
- Use `figsize=(6, 4)` — fits a letter-sized page with 1" margins
- Set font size to match your document (typically 11 pt for publications)
- Apply `tick_params(which='both', direction='in', top=True, right=True)` for publication plots
- Use `bbox_inches='tight'` when saving to avoid cropped labels
- Save as PDF or SVG for papers; PNG at 300 dpi for theses

**Code organization**
- Put all imports and `rcParams.update()` in the first code cell
- Write reusable plotting functions with docstrings for repeated plot types
- Use `fig, ax = plt.subplots()` (object-oriented interface) for any plot with more than one panel or a colorbar
